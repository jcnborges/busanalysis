```python
import MAG as mag
import sub_algorithms as sub

print('Creation of MAG')
G = mag.MultiAspectGraph()
print('Mag order: {}'.format(G.order()))

G.add_edge(('A', 2),('B', 2), weight=0)
G.add_edge(('A', 3),('B', 0), weight=1)
G.add_edge(('A', 4),('B', 1), weight=23)
G.add_edge(('A', 4),('B', 2), weight=31)
G.add_edge(('A', 3),('B', 0), weight=34)
G.add_edge(('A', 4),('B', 1), weight=10)
G.add_edge(('A', 5),('B', 3), weight=4)
G.add_edge(('B', 4),('A', 5), weight=4)
e = [(('A', 2),('B', 4), {'weight':5}), (('C', 2),('B', 2), {'weight':90}),
     (('A', 2),('B', 5), {'weight':5}), (('A', 2),('B', 6), {'weight':5}),
     (('A', 2),('B', 4), {'weight':14})]
G.add_edges_from(e)

print('Mag order: {}'.format(G.order()))
print('Edges')
print(G.edges())

print('Visited edges by DFS')
edges = sub.sub_dfs_edges(G, [1,0], source=('A',))

for e in edges:
    print(e)

H = sub.sub_dfs_tree(G, [1,0])
print(type(H))
print(H.edges())

H = sub.sub_dfs_tree(G, [1,0], ('B',))
print(type(H))
print(H.edges())

print('Predecessors')
print(sub.sub_dfs_predecessors(G,[1,0]))
print(sub.sub_dfs_predecessors(G,[0,1]))

print('Sucessors')
print(sub.sub_dfs_successors(G,[1,0]))

print('Visited edges by dfs')
print(sub.sub_bfs_edges(G,[1,0], ('A',)))

H = sub.sub_bfs_tree(G, [1,0], ('B',))
print(type(H))
print(H.edges())

print('Predecessors - bfs')
print(sub.sub_bfs_predecessors(G, [1,0], ('A',)))

print('Dijstra path')
print(sub.sub_dijkstra_path(G, [1,0], ('A',), ('B',), weight='weight'))
```
